```python
import cv2
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt
```

```python
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
```

```python
img = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)
```

```python
target_size = (128, 128)
resized_img = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)
```

```python
normalized_img = resized_img.astype('float32') / 255.0
```

```python
plt.imshow(normalized_img, cmap='gray')
plt.title(f"Processed Image Shape: {normalized_img.shape}")
plt.axis('off')
plt.show()
```

```python
def relu_activation(x):
    """
    ReLU funksiyasi: musbat qiymatlarni o'tkazadi,
    manfiylarini 0 ga aylantiradi.
    """
    return np.maximum(0, x)
```

```python
# Lekin model ichida weights va biaslar bilan ReLU muhim rol o'ynaydi.
activated_img = relu_activation(normalized_img)

print("Jarayon yakunlandi. Rasm o'lchami:", normalized_img.shape)
```
