1. Interpreter o'zi nima?
Python — interpretatsiya qilinadigan til. Bu degani, siz yozgan kodni kompyuter tushunadigan "0" va "1"larga aylantirish uchun maxsus dastur (Interpreter) kerak bo'ladi. Siz kod yozasiz, interpreter uni qatorma-qatar o'qiydi va darhol bajaradi.

Argumentlarni uzatish (Argument Passing)

Siz skriptni ishga tushirayotganda unga qo'shimcha ma'lumotlar (argumentlar) berib yuborishingiz mumkin. Python bularni sys.argv ro'yxatiga (list) joylaydi.

    sys.argv[0] — skriptning nomi.

    sys.argv[1] — birinchi argument va hokazo.


1. PATH nima?

PATH — bu operatsion tizimingizdagi (Windows, macOS, Linux) muhit o'zgaruvchisi (environment variable). Uni kompyuteringizning "qidiruv xaritasi" deb tasavvur qiling.
2. sys moduli qanday ishlaydi?

sys (system) moduli Python interpreterining o'zi bilan va u ishlayotgan muhit (operatsion tizim) bilan muloqot qilish uchun ishlatiladi. Bu modul Python o'rnatilganda doimo birga keladi.

Mana eng ko'p ishlatiladigan qismlari:
sys.argv (Argumentlar)

Skriptni ishga tushirayotganda tashqaridan ma'lumot uzatish uchun kerak.
