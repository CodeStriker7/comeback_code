```python
print("hello colab")
tasks =[]
def add(name):
	item = {"name": name, "done": False}
	tasks.append(item)
	print(f"Added: {name}")

def show():
	print("\n--- Tasks List ---")
	for i, t in enumerate(tasks):
		status = "V" if t["done"] else " "
		print(f"{i}. [{status}] {t['name']}")

def check(id):
	if 0 <= id < len(tasks):
		tasks[id]["done"] = True
		print(f" Tasks {id} is done!")
	else:
		print("Error invalid ID")

def clear():
	tasks.clear()
	print("list is empty")

add("learning python with brain and code editor sublime text")
add("second Added")
show()
check(0)
check(1)
clear()
```
