```python
import numpy as np
import pandas as pd


np.random.seed(42)
n_samples = 50

ram = np.random.choice([4, 6, 8, 12, 16], n_samples)
storage = np.random.choice([64, 128, 256, 512], n_samples)
screen_size = np.round(np.random.uniform(5.5, 6.9, n_samples), 1)
price = (ram * 45) + (storage * 0.4) + (screen_size * 15) + np.random.randint(20, 100, n_samples)

data = np.column_stack((ram, storage, screen_size, price))
# pandas bilan csv ga utkazamiz
columns = ['ram (GB)', 'storage (GB)', 'screen_size (dyum)', 'price ($)']
df = pd.DataFrame(data, columns=columns)

df.to_csv('smartfon_data.csv', index=False)
print("Dataset 'smartfon_data.csv' nomi bilan saqlandi")


```
